```python
from bs4 import BeautifulSoup
import requests as rq
```

```python
url = 'http://www.tribonet.org/keeping-the-wind-in-your-sails/'
```

```python
content = rq.get(url)
```

```python
soup = BeautifulSoup(content.text, 'html.parser')
```

```python
for tag in soup.find_all(['script', 'style']):
    tag.decompose()
```

```python
print(soup.prettify())
```

```python
print(soup.get_text())
```
